#  상위 50개 데이터만 뽑아서 시각화하기

```python
# 직접 리콜 사유와 관련이 적은 문법적 어구들을 배제하기 - stopwords
spwords = set(["동안", "인하여", '있는', '경우', '있습니다', '가능성이', '않을', '차량의',
              '가', '에', '될', '이', 
              '인해', '수', '중', '시', '또는', '있음', '의','있다', '및', '있으며',
              '발생할', '이로', '오류로', '해당'])

# 리콜 사유에 해당하는 열의 값들을 중복 제거한 뒤 모두 이어붙여 text라는 문자열로 저장하기
text = ""
for c in df.리콜사유.drop_duplicates():
    text += c
text[:100]
# '제작 공정 중 라디에이터 캡의 제조 불량으로 라디에이터 캡과 필러넥이 제대로 밀착되지 않아 라디에이터 내부 압력을 일정하게 유지할 수 없습니다.
# 이로 인해 라디에이터 내부 압력이 '
```

```python
# 한글 사용 전 폰트 지정
wc1 = WordCloud(max_font_size=200, stopwords=spwords,
               font_path=r'C:/work/cro1/나눔 글꼴/나눔고딕/NanumFontSetup_OTF_GOTHIC/NanumGothic.otf',
               background_color='white', width=800, height=800)
wc1.generate(text)


plt.figure(figsize=(10,5))
plt.imshow(wc1)
plt.tight_layout(pad=0)
plt.axis('off')
plt.show()
```
